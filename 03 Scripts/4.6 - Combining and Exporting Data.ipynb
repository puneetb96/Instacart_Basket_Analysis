## 1. Creating Data

```python
# Define a dictionary containing January 2020 data
data1 = {'customer_id': ['6732','767','890','635'],
         'month':['Jan-20','Jan-20','Jan-20','Jan-20'],
         'purchased_meat':[0,13,3,4],
         'purchased_alchol':[1,2,10,0],
         'purchased_snacks':[10,5,1,7]}
```

```python
# Define a dictionary containing February 2020 data
data2 = {'customer_id': ['6732','767','890','635'],
         'month':['Feb-20','Feb-20','Feb-20','Feb-20'],
         'purchased_meat':[0,10,5,3],
         'purchased_alchol':[2,4,14,0],
         'purchased_snacks':[10,5,1,7]}
```

```python
#Convert the dictionary into dataframe (pd.DataFrame function)
df = pd.DataFrame(data1,index = [0,1,2,3])
df_1 = pd.DataFrame(data2,index = [0,1,2,3])
```

1. pd.DataFrame function turns dictionary into dataframe
2. index argument tells Pythong how to be indexed. Rows are 0-3 as listed below

```python
#January dataframe
df
```

```python
#February dataframe
df_1
```

## 2. Concatenate Data

1. First create list of frames
2. Then use pd.concat() function to concatenate frames

```python
#Create frames
frames = [df,df_1]
```

```python
#Concat frames
df_concat = pd.concat(frames)
```

```python
#New concatenated dataframe
df_concat
```

- Default argument in play. 'axis' argument
- default is set to "long" which is axis = 0
- axis = 1, is "wide" stack, side by side concatenation

```python
#Wide stack example, axis = 1
df_concat = pd.concat(frames, axis = 1)
```

```python
#New wide look
df_concat
```

- To recap, the pd.concat() function:

1. Is suitable for rows or columns of the same length.
2. Will place dataframes on top of each other by default (axis = 0).
3. Requires a list as its main argument (this is why the frames list was created first in the example above).

## Joining and Merging Data

```python
#df.join() and df.merge() functions
```

- Won't be exploring joins
- Article of Pandas joins vs. merge article for more info

- df.merge() function is best when dataframes don't match in shapes
- Need a key or common identifier to match the datasets

```python
#Create new third dataframe
data3 = {'customer_id': ['6732','767','890','635'],
         'month':['Jan-20','Jan-20','Jan-20','Jan-20'],
         'days_purchased_on':[0,10,4,1]}
```

```python
#Create new data frame 3
df_2 = pd.DataFrame(data3, index = [0,1,2,3])
```

```python
#New dataframe
df_2
```

```python
#Merge function
df_merged = df.merge(df_2, on = 'customer_id')
```

```python
#Merged dataframe
df_merged
```

- df.merge() function took first dataframe (df) and combined with df_2 "on" customer_id column
- Might need to rename columns in some instances "cust_id" vs "customer_id"

Month x and month y showing. Fix this by adding month as a key in function like below

```python
#Adding month key
df_merged = df.merge(df_2, on = ['customer_id','month'])
```

```python
#Check results
df_merged
```

Checking for full math on data with "indicator = True" argument

```python
#Indicator Argument
df_merged = df.merge (df_2, on = ['customer_id','month'], indicator = True)
```

```python
df_merged
```

- "both" both means full match between datasets
- "left_only" or "right_only" indicates only existing in one dataframe

```python
#Check value counts
df_merged['_merge'].value_counts()
```

"_merge" also known as the merge flag

```python
#Test merge without overwriting
pd.merge(df,df_2, on = ['customer_id', 'month'], indicator = True)
```

- When doing test merge, include both dataframes in parenthesis like example above

- "how" argument tells you what kind of merge()
- inner(default), left,right, outer
- Ex: df.merge (df_2, on = ['customer_id','month'], how = 'inner')

## Exploring Data in Pickle Format

- "Pickle" format is .pkl instead of .csv
- CSV used in multiple tools (SAS, Excel, R), but are much larger and take longer, leads to index column issues when importing
- PKL files are super quick and work well in Jupyter, but only work in Python. NOt assecible with Excel, etc.

## Export Merged File with both CSV and PKL Formats

```python
#CSV Export
df_merged_large.to_csv(os.path.join(path, '02 Data','Prepared Data','orders_products_combined.csv'))
```

```python
#PKL Export
df_merged_large.to_pickle(os.path.join(path, '02 Data','Prepared Data','orders_products_combined.pkl'))
```
