## 04. Subsetting

- Creating smaller dataset from a whole dataset (df_prods) based on a filter
- We're creating subset of df_prods that only contains data from snacks department

```python
df_snacks = df_prods[df_prods['department_id']==19]
```

```python
#Indexing element breakdown
df_prods['department_id']==19
```

This shows you what Python is looking at. Showing all values and assinging True or False based on if it's in the snacks department

```python
#This next line wraps the initial dataframe to the logic we just wrote above 
#Now this dataframe only shows true values of department 19 values
df_prods[df_prods['department_id']==19]
```

We labled this earlier as df_snacks now

```python
#First 5 rows of df_snacks, our new snacks only dataframe
df_snacks.head()
```

Can also use the 'loc' function to get same result

```python
#loc function example 1
df_snacks_2 = df_prods.loc[df_prods['department_id']==19]
```

```python
#loc functione xample 2, you can check multiple values at the same time (.isin([17,18,19]))
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```
