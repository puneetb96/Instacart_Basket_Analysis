The 'product_id' column is the only shared key by both dataframes. We want to do an inner join bcause we're only looking for data that matches both tables

6. Confirm the results of the merge using the merge flag

```python
#Test merge before overwriting, with merge flag
pd.merge(df_merged_large, df_prod, on = 'product_id', how = 'inner', indicator = True)
```

```python
#Actual merge
ords_prods_merge = df_merged_large.merge(df_prod, on = 'product_id', how = 'inner', indicator = True)
```

```python
ords_prods_merge['_merge'].value_counts()
```
